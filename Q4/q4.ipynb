```python
def do():
    import csv
    f = open('q4.csv')
    data = csv.reader(f)
    header = next(data)

    maxsta_l1=0
    maxsta_l2=0
    maxsta_l3=0
    maxsta_l4=0
    minsta_l1=0
    minsta_l2=0
    minsta_l3=0
    minsta_l4=0
    max_sum = [0,0,0,0]
    min_sum = [999999999,999999999,999999999,999999999]

    for row in data:
        check = 0
        cursum = 0
        for i in range(4,6):
            if(row[i] == ''):
                check = 1
                continue
            else:
                row[i] = float(row[i])
        if (check == 0) :
            if(row[1]=="1호선"):
                cursum += row[4]
                cursum += row[5]
                if(max_sum[0] < cursum):
                    max_sum[0] = cursum
                    maxsta_l1 = row[3]
                if(min_sum[0] > cursum):
                    min_sum[0] = cursum
                    minsta_l1 = row[3]
            elif(row[1]=="2호선"):
                cursum += row[4]
                cursum += row[5]
                if(max_sum[1] < cursum):
                    max_sum[1] = cursum
                    maxsta_l2 = row[3]
                if(min_sum[1] > cursum):
                    min_sum[1] = cursum
                    minsta_l2 = row[3]
            elif(row[1]=="3호선"):
                cursum += row[4]
                cursum += row[5]
                if(max_sum[2] < cursum):
                    max_sum[2] = cursum
                    maxsta_l3 = row[3]
                if(min_sum[2] > cursum):
                    min_sum[2] = cursum
                    minsta_l3 = row[3]
            elif(row[1]=="4호선"):
                cursum += row[4]
                cursum += row[5]
                if(max_sum[3] < cursum):
                    max_sum[3] = cursum
                    maxsta_l4 = row[3]
                if(min_sum[3] > cursum):
                    min_sum[3] = cursum
                    minsta_l4 = row[3]
    
    print("*** Subway Report for Seoul on March 2023 ***")
    print("Line1:")
    print("Busiest Station:",maxsta_l1,"(",int(max_sum[0]),")")
    print("Least used Station :",minsta_l1,"(",int(min_sum[0]),")")
    print("Line2:")
    print("Busiest Station:",maxsta_l2,"(",int(max_sum[1]),")")
    print("Least used Station :",minsta_l2,"(",int(min_sum[1]),")")
    print("Line3:")
    print("Busiest Station:",maxsta_l3,"(",int(max_sum[2]),")")
    print("Least used Station :",minsta_l3,"(",int(min_sum[2]),")")
    print("Line4:")
    print("Busiest Station:",maxsta_l4,"(",int(max_sum[3]),")")
    print("Least used Station :",minsta_l4,"(",int(min_sum[3]),")")
    
    f.close()   
    
if __name__ == '__main__':
    do()
```
